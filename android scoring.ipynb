```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
```

```python
origin = pd.read_csv('mbll_2019-08-17_172235.722.txt')
origin = origin.rename(columns = {origin.columns[0]: 'Time'})
```

```python
file_name = "mbll_2019-08-17_172235.722.txt"

# date_info
date_info = file_name.split("_")[1][:10]
year = date_info[:4]
month = date_info[5:7]
day = date_info[8:]
date_info_str = year + "년 " + month + "월 " + day + "일"

# time_info
time_info = file_name.split(".")[0][-8:]
hour = time_info[:2]
minute = time_info[3:5]
time_info_str = hour + "시 " + minute + "분 "
```

```python
origin_sliced = origin[origin.Time >= 6]

# moving average
movinglpf = origin_sliced.rolling(window=10).mean()
movinglpf['Time']=origin_sliced['Time']
movinglpf = movinglpf[origin.Time >= 8]
movinglpf
```

```python
# HbR 컬럼 다 삭제하기
# for i in range(16, 32):
#     del movinglpf["HbR."+str(i)]

if (movinglpf.shape[1] == 18):
    del movinglpf[movinglpf.columns[17]]

# ' Average '라는 컬럼 만들기
movinglpf['Average'] = movinglpf['0']
for i in range(1, 16):
    movinglpf['Average'] = movinglpf['Average'] + movinglpf['0.'+str(i)]
movinglpf['Average'] = movinglpf['Average']/ 16

movinglpf
```

```python
np_movinglpf = np.array(movinglpf)
np_time = np_movinglpf[:, 0]
np_time
```

```python
# 시간에 따라 구간별로 자르기
def splitTask(np_data):
    split_list = []
    
    quotient = (np_time[-1] - np_time[0])//25
    residual = (np_time[-1] - np_time[0])%25
    
    minTime = 8
    data = np_data[:, :] # shalllow copy
    for idx in range(int(quotient)):
        maxTime = minTime + 25
        min_idx = np.where(np_time>=minTime)[0][0]
        max_idx = np.where(np_time<maxTime)[0][-1]
        output = data[:, :][min_idx:max_idx+1]
        split_list.append(np.array(output))
        minTime = maxTime
        
    # 나머지 
    if residual >= 10:
        min_idx = np.where(np_time>=minTime)[0][0]
        max_idx = len(np_time)-1
        output = data[:, :][min_idx:max_idx+1]
        split_list.append(np.array(output))
        
    return split_list
```

```python
split_list = splitTask(np_movinglpf)
```

# 구간마다 평균, 기울기 구하기

```python
# task별로 25초씩 slope, average 구하기
from scipy.optimize import curve_fit

def func(x, a, b):
    return a*x + b
```

```python
taskNum = len(split_list)

avg_list = []
slope_list = []


for i in range(16,32):
    avg_temp = []
    slope_temp = []
    for j in range(taskNum):
        x = split_list[j][:, 0]
        y = split_list[j][:, i-16]
        popt, pcov = curve_fit(func, x, y)
        slope_temp.append(popt[0])
        avg_temp.append(np.average(y))

    avg_list.append(avg_temp)
    slope_list.append(slope_temp)
    
avg_list
```

# 점수 구하기

```python
# 채널별 score
score_2list = []

for i in range(16):
    score_list = []
    min_slope = 1000
    min_avg = 1000
    for j in range(1, taskNum):
        
        # 10점
        if (avg_list[i][0] <= avg_list[i][j] and slope_list[i][j] >= 0):
            score_list.append(10)
            if (avg_list[i][j] < min_avg): # 10점 중에 가장 작은 avg 값
                min_avg = avg_list[i][j]
            if (slope_list[i][j] < min_slope):# 10점 중에 가장 작은 slope 값
                min_slope = slope_list[i][j]
        
        # 7점
        elif (avg_list[i][0] <= avg_list[i][j] and slope_list[i][j] < 0):
            score_list.append(7)

        # 4점
        elif (avg_list[i][0] > avg_list[i][j] and slope_list[i][j] >= 0):
            score_list.append(4)
            
        # 1점
        elif (avg_list[i][0] > avg_list[i][j] and slope_list[i][j] < 0):
            score_list.append(1)
    
    for j in range(len(score_list)):
        if (score_list[j] == 7 and avg_list[i][j] > min_avg):
            score_list[j] = 8
        if (score_list[j]==4 and slope_list[i][j] > min_slope):
            score_list[j] = 5
            
    score_2list.append(score_list)
score_2list
```

```python
# 최종 score
final_score = np.average(score_2list) * 10
final_score
```
